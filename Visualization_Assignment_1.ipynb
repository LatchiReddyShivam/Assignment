##Q5)Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.




##Ans- **Steps:**



1.   **Generate Correlated Data:** Use NumPy to generate a dataset with correlated features.

2.   **Calculate Correlation Matrix:** Use np.corrcoef() to calculate the correlation between features.

3.   **Create Heatmap:** Use Seabornâ€™s heatmap() to visualize the correlation matrix.





```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Generate correlated data
data = np.random.multivariate_normal([0, 0, 0],
                                      [[1, 0.8, 0.5],
                                       [0.8, 1, 0.3],
                                       [0.5, 0.3, 1]],
                                      size=1000)

# Create DataFrame
df = pd.DataFrame(data, columns=['Feature 1', 'Feature 2', 'Feature 3'])

# Calculate the correlation matrix
correlation_matrix = df.corr()

# Create a heatmap
plt.figure(figsize=(8, 6))  # Optional: specify figure size
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")

# Add title
plt.title("Correlation Matrix Heatmap")

# Display the plot
plt.show()
```

1.   **Generate X and Y Data:** Create a grid of values for X and Y.

2.   **Calculate Z Values:** Compute z based on the given function.

3.   **Create 3D Surface Plot:** Use Plotlyâ€™s surface() to visualize the function.






```python
import numpy as np
import pandas as pd
import plotly.express as px

# Generate x and y data
x = np.linspace(-5, 5, 100)  # x values from -5 to 5
y = np.linspace(-5, 5, 100)  # y values from -5 to 5

# Create a meshgrid from x and y
x, y = np.meshgrid(x, y)  # Create a grid of x and y values

# Calculate z values based on the function z = sin(âˆš(xÂ² + yÂ²))
z = np.sin(np.sqrt(x**2 + y**2))

# Create a DataFrame
data = {
    'X': x.flatten(),  # Flatten x values
    'Y': y.flatten(),  # Flatten y values
    'Z': z.flatten()   # Flatten z values
}
df = pd.DataFrame(data)  # Create a DataFrame from the flattened data

# Create a 3D surface plot using Plotly
fig = px.imshow(z, x=x[0], y=y[:,0], title='3D Surface Plot of z = sin(âˆš(xÂ² + yÂ²))',
                 color_continuous_scale='Viridis', aspect='auto')

# Update layout for better display
fig.update_layout(scene=dict(
        xaxis_title='X Axis',
        yaxis_title='Y Axis',
        zaxis_title='Z Axis'),
    title='3D Surface Plot of z = sin(âˆš(xÂ² + yÂ²))',
    autosize=True)

# Show the plot
fig.show()
```

##Q1)Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.





##Ans- **Steps:**



1.   **Import Required Libraries:** Import Bokeh libraries along with NumPy for calculations.

2.   **Generate Data:** Create x-values from 0 to 10 and calculate y-values as the sine of x.

3.   **Create the Plot:** Use Bokehâ€™s figure and line functions to create and display the sine wave plot.



```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Output in Jupyter Notebook
output_notebook()

# Generate data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a Bokeh plot
p = figure(title="Sine Wave", x_axis_label='X', y_axis_label='sin(X)')
p.line(x, y, legend_label='sin(x)', line_width=2, color='blue')

# Show the plot
show(p)
```

##Ans- **Steps:**


1.   **Generate Data:** Create random normal data.


2.   **Create Histogram:** Calculate the histogram and use Bokehâ€™s quad() to visualize it.



```python
# Generate random data
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

# Create a Bokeh histogram
p = figure(title="Histogram of Random Data", x_axis_label='Value', y_axis_label='Frequency')
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color='navy', line_color='white', alpha=0.5)

# Show the plot
show(p)
```
