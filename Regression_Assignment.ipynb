##Q1)What is Simple Linear Regression?
##Ans- Simple Linear Regression is a statistical technique used to model the relationship between two variables: one independent variable (X) and one dependent variable (Y). The goal is to predict the value of Y based on X by fitting a straight line to the data. This line is expressed as Y=mX+c, where ùëö represents the slope (rate of change) and ùëê is the y-intercept (value of ùëå when ùëã=0). It is widely used in forecasting and establishing relationships in data.


##Q2)What are the key assumptions of Simple Linear Regression?

##Ans-



##1.   **Linearity**: The relationship between the independent and dependent variables is linear. A scatter plot of ùëã and ùëå should form a straight-line pattern.

##2.   **Independence**: Observations in the dataset are independent of each other. No patterns like time-based dependencies should exist.

##3.   **Homoscedasticity**: The variance of residuals (errors) should be constant across all levels of ùëã. This ensures that predictions are consistent.

##4.   **Normality**: Residuals should be normally distributed, which can be checked using Q-Q plots or statistical tests. This assumption ensures valid hypothesis testing and confidence intervals.

##Q3)What does the coefficient ùëö represent in the equation ùëå = ùëöùëã+ùëê?

##Ans-The slope ùëö quantifies how much ùëå changes for a one-unit increase in ùëã. For example, in predicting house prices based on area, if ùëö = 2000, it means the price increases by $2000 for each additional square foot of area. The sign of ùëö indicates the direction: positive ùëö means a direct relationship, and negative ùëö implies an inverse relationship.

##Q4)What does the intercept ùëê represent in the equation ùëå = ùëöùëã+ùëê?

##Ans-The intercept ùëê is the value of ùëå when ùëã = 0. It provides the baseline prediction in the absence of any influence from ùëã. For example, if ùëê = 5000 in the house price model, it means the starting price of a house, even with zero area, is $5000. While often not meaningful in real-world contexts (as zero ùëã may not be realistic), it helps in defining the regression line mathematically.

##Q8)What is Multiple Linear Regression?


##Ans-Multiple Linear Regression extends Simple Linear Regression to model the relationship between one dependent variable $X_1$,$X_2$,...,$X_k$. Its equation is:
##Y = $b_0$ + $b_1$$X_1$ + $b_2$$X_2$ +...+ $b_k$$X_k$

##Here, $b_0$ is the intercept, and $b_1$,$b_2$,...$b_k$ are the coefficients representing the contribution of each predictor.


##Q9)What is the main difference between Simple and Multiple Linear Regression?

##Ans-Simple Linear Regression involves one independent variable, while Multiple Linear Regression uses two or more predictors. The latter captures more complex relationships, allowing for better predictions when multiple factors influence the dependent variable.


##Q10)What are the key assumptions of Multiple Linear Regression?

##Ans-

##1.   Linearity: The relationship between predictors and the dependent variable is linear.

##2.   No multicollinearity: Predictors should not be highly correlated with each other.

##3.   Homoscedasticity: The variance of residuals should be constant across all levels of predictors.

##4.   Normality: Residuals should follow a normal distribution.

##5.   Independence: Observations must be independent.


##Q11)What is heteroscedasticity, and how does it affect the results of a Multiple Linear Regression model?


##Ans-Heteroscedasticity occurs when the variance of residuals is not constant across all levels of the independent variables. This violates one of the key assumptions of regression.



##1.   **Impact:** It leads to inefficient estimates, inflated standard errors, and unreliable hypothesis tests, potentially distorting the significance of predictors.

##2.   **Example:** In financial data, residuals often increase as the value of the independent variable (e.g., income) increases, indicating heteroscedasticity.


##Q12)How can you improve a Multiple Linear Regression model with high multicollinearity?


##Ans-

##1.   **Remove Correlated Variables:** Exclude one of the highly correlated predictors based on domain knowledge.

##2.   **Regularization:** Use Ridge or Lasso regression to penalize large coefficients and reduce multicollinearity.

##3.   **Principal Component Analysis (PCA):** Transform predictors into uncorrelated components.

##4.   **Variance Inflation Factor (VIF):** Identify predictors with high VIF scores and address them.

##Q23)What is polynomial regression?

It extends linear regression to capture non-linear relationships by including higher-order terms of ùëã. The general form is:
##ùëå =$b_0$ + $b_1$$X$ + $b_2$$X^2$ +...+ $b_n$$X^n$



##Q24)How does polynomial regression differ from linear regression?


##Ans-

##1.   **Linear Regression** assumes a straight-line relationship between ùëã and ùëå, using only ùëã as a predictor.

##2.   **Polynomial Regression** allows for curves by including higher-degree terms ($X^2$, $X^3$,...)

##**Example:** A quadratic model (ùëå =$b_0$ + $b_1$$X$ + $b_2$$X^2$)can fit a parabolic trend, which a linear model cannot.



##Q25)When is polynomial regression used?


##Ans-Polynomial regression is used when data exhibits a non-linear relationship that cannot be captured by a simple linear model. It is commonly applied in fields such as physics, economics, and biology to model complex phenomena like growth trends or diminishing returns.

##Q28)What are the limitations of polynomial regression?

##Ans-

##2.   **Complexity:** Higher-degree models can be computationally intensive and difficult to interpret.

##3.   **Extrapolation Issues:** Polynomial models can produce unrealistic predictions outside the observed range of ùëã.

##Q31)How is polynomial regression implemented in Python?



##Ans- Polynomial regression can be implemented using **PolynomialFeatures** from **sklearn.preprocessing** to transform the data into polynomial features, followed by fitting a linear regression model. Here is an example:
















```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt

# Generate synthetic data
X = np.linspace(-3, 3, 100).reshape(-1, 1)  # X is 2D for sklearn
y = (2 * X**3 - 5 * X**2 + 4 * X + np.random.randn(100, 1)).ravel()  # y is 1D for sklearn

# Transform data to include polynomial features
poly = PolynomialFeatures(degree=3)
X_poly = poly.fit_transform(X)

# Fit model
model = LinearRegression()
model.fit(X_poly, y)

# Predict and visualize
y_pred = model.predict(X_poly)

# Plot the results
plt.scatter(X, y, color='blue', label='Data points')  # Scatter plot for actual data
plt.plot(X, y_pred, color='red', label='Polynomial fit')  # Polynomial regression line
plt.title('Polynomial Regression')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```
