##Q1: What is a Support Vector Machine (SVM)?



##Ans- A Support Vector Machine (SVM) is a supervised learning algorithm used for classification and regression tasks. It works by finding the optimal hyperplane that best separates the classes in the feature space. The main objective of SVM is to maximize the margin, which is the distance between the separating hyperplane and the closest data points (support vectors).



##Q2: What is the difference between Hard Margin and Soft Margin SVM?



##Ans-

##1.   **Hard Margin SVM** strictly separates the classes with a hyperplane and does not allow any misclassification. It works well for linearly separable data but fails with noisy data.


##2.   **Soft Margin SVM** allows some misclassification by introducing a penalty term (C parameter) to handle noise and overlapping data points, making it more flexible for real-world scenarios.



##Q3: What is the mathematical intuition behind SVM?


##Ans- SVM optimizes a hyperplane by solving the following optimization problem:

##$$ \min \frac{1}{2} ||w||^2 $$
##subject to  
##$$ y_i (w \cdot x_i + b) \geq 1 $$


##for all training samples $(x_i, y_i) $. Here, $w$ is the weight vector,$ùëè$ is the bias term, and $ùë¶ùëñ$ represents class labels. The margin is maximized to ensure better generalization.




##Q4: What is the role of Lagrange Multipliers in SVM?


##Ans-Lagrange multipliers are used to convert the constrained optimization problem into an unconstrained dual optimization problem. They help in solving the problem efficiently using quadratic programming and enable the use of the kernel trick for non-linear data.



##Q5: What are Support Vectors in SVM?


##Ans- Support vectors are the data points that lie closest to the decision boundary. They are crucial because they determine the position and orientation of the optimal hyperplane. Removing any other points does not affect the decision boundary.


##Q6: What is a Support Vector Classifier (SVC)?


##Ans- A Support Vector Classifier (SVC) is an extension of SVM used for classification tasks. It finds the best hyperplane that separates different classes while maximizing the margin and minimizing classification errors.


##Q7: What is a Support Vector Regressor (SVR)?


##Ans-A Support Vector Regressor (SVR) is an adaptation of SVM for regression tasks. Instead of finding a hyperplane for classification, it finds a function that fits the data within a margin of tolerance (epsilon).



##Q8: What is the Kernel Trick in SVM?


##Ans- The Kernel Trick allows SVM to handle non-linearly separable data by transforming the original input space into a higher-dimensional feature space where a linear separation is possible. Common kernels include:




##1.   Linear Kernel: $K$($x,y$) = $x.y$



##2.  **Polynomial Kernel:**

## $K(x, y) = (x \cdot y + c)^d$



##3.   **RBF Kernel (Radial Basis Function):**

##$K(x, y)$ = $( e^{-\gamma}(x - y)^2)$



##Q9: Compare Linear Kernel, Polynomial Kernel, and RBF Kernel.


##Ans-




##1.   The **Linear Kernel** is most suitable for datasets that are linearly separable, as it efficiently separates classes using a straight hyperplane.


##2.   The **Polynomial Kernel** is useful when dealing with complex relationships between features, as it captures interactions between different input features by transforming them into higher dimensions.


##3.   The **RBF (Radial Basis Function) Kernel** is highly flexible and works well for datasets that are not linearly separable, as it maps data into an infinite-dimensional space where a linear separation can be achieved.


##Q10: What is the effect of the C parameter in SVM?


##Ans- The C parameter controls the trade-off between margin size and misclassification.



##1.   A **small C** allows a larger margin but tolerates more misclassification.

##2.   A **large C** prioritizes correct classification but may lead to overfitting.



##Q11: What is the role of the Gamma parameter in RBF Kernel SVM?


##Ans- The gamma parameter in the RBF kernel determines the influence of individual training points.

##1.   **Low gamma** means points far apart influence each other, resulting in a smooth decision boundary.

##2.   **High gamma** means each point has more localized influence, possibly leading to overfitting.



##Q12: What is the Na√Øve Bayes classifier, and why is it called "Na√Øve"?



##Ans- Na√Øve Bayes is a probabilistic classifier based on Bayes' Theorem. It assumes that features are **independent** given the class label, which is often unrealistic, hence the term ‚Äúna√Øve.‚Äù Despite this assumption, it works well in many applications.


##Q13: What is Bayes‚Äô Theorem?

##Ans- Bayes' Theorem provides a way to update prior knowledge with new evidence:

##$$ P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B)} $$

##where:


##1.   P(A|B) is the posterior probability

##2.   P(B|A) is the likelihood

##3.   P(A) is the prior probability

##4.   P(B) is the evidence



##Q14: Explain the differences between Gaussian Na√Øve Bayes, Multinomial Na√Øve Bayes, and Bernoulli Na√Øve Bayes.



##Ans-

##1.   Gaussian Na√Øve Bayes assumes that numerical features follow a normal (Gaussian) distribution, making it ideal for continuous data such as age, temperature, or height.

##2.   Multinomial Na√Øve Bayes is typically used for discrete count data, such as word frequency in text classification problems like spam detection.


##3.   Bernoulli Na√Øve Bayes, on the other hand, is designed for binary feature data, where each feature is either present or absent, making it useful for applications like document classification based on word occurrence.



##Q15: When should you use Gaussian Na√Øve Bayes over other variants?



##Ans- Use Gaussian Na√Øve Bayes when the feature values are continuous and approximately normally distributed, such as in medical datasets or sensor readings.



##Q16: What are the key assumptions made by Na√Øve Bayes?



##Ans-

##1.   Features are **conditionally independent** given the class label.

##2.   All features contribute equally to the classification.

##3.   Data follows a known distribution (Gaussian, multinomial, or Bernoulli).




##Q17: What are the advantages and disadvantages of Na√Øve Bayes?



##Ans- **Advantages:**

##1.   Fast and efficient for large datasets.

##2.   Works well with small datasets and limited training data.

##3.   Performs well in text classification problems.

##**Disadvantages:**

##1.   Assumes feature independence, which is often unrealistic.

##2.   Struggles with highly correlated features.




##Q18: Why is Na√Øve Bayes a good choice for text classification?



##Ans-

##1.   Works well with high-dimensional data.

##2.   Efficient for real-time applications.

##3.   Handles class imbalances effectively.




##Q19: Compare SVM and Na√Øve Bayes for classification tasks.



##Ans-

##1.   SVM is a robust algorithm that works well for high-dimensional data and complex decision boundaries. It is particularly effective when the margin between classes is well-defined. However, it can be computationally expensive, especially for large datasets.


##2.   Na√Øve Bayes, in contrast, is a simple and efficient algorithm that works well for text classification tasks. It assumes feature independence, which simplifies computation, but this assumption may not always hold in real-world scenarios.



##Q20: How does Laplace Smoothing help in Na√Øve Bayes?


##Ans- Laplace Smoothing prevents zero probabilities by adding a small value (ùõº) to each term in the probability calculation, ensuring unseen words or features do not lead to a probability of zero.

##Q30: Write a Python program to train a Gaussian Na√Øve Bayes model and compare the predictions before and after Laplace Smoothing.

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler

# Step 1: Load the Iris dataset
data = load_iris()
X = data.data
y = data.target

# Step 2: Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Step 3: Train a Gaussian Naive Bayes model without Laplace smoothing
gnb_no_smoothing = GaussianNB(var_smoothing=1e-9)  # setting very small variance smoothing as default
gnb_no_smoothing.fit(X_train, y_train)

# Step 4: Predict and evaluate the model without Laplace smoothing
y_pred_no_smoothing = gnb_no_smoothing.predict(X_test)
accuracy_no_smoothing = accuracy_score(y_test, y_pred_no_smoothing)

print(f"Accuracy without smoothing: {accuracy_no_smoothing:.4f}")

# Step 5: Apply Laplace smoothing by setting higher value for var_smoothing
gnb_laplace_smoothing = GaussianNB(var_smoothing=1.0)  # A higher value applies more smoothing
gnb_laplace_smoothing.fit(X_train, y_train)

# Step 6: Predict and evaluate the model with Laplace smoothing
y_pred_laplace_smoothing = gnb_laplace_smoothing.predict(X_test)
accuracy_laplace_smoothing = accuracy_score(y_test, y_pred_laplace_smoothing)

print(f"Accuracy with Laplace smoothing: {accuracy_laplace_smoothing:.4f}")
```

##Q33: Write a Python program to implement a Na√Øve Bayes classifier for spam detection using email data.

```python
import pandas as pd
import zipfile
import os
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import urllib.request

# Step 1: Download and extract the dataset
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/00228/smsspamcollection.zip"
zip_file = "smsspamcollection.zip"
dataset_folder = "smsspamcollection"

# Download the dataset
urllib.request.urlretrieve(url, zip_file)

# Extract the zip file
with zipfile.ZipFile(zip_file, 'r') as zip_ref:
    zip_ref.extractall(dataset_folder)

# Step 2: Load the dataset
# The dataset is now in the extracted folder, and the file is called SMSSpamCollection
dataset_path = os.path.join(dataset_folder, "SMSSpamCollection")
dataset = pd.read_csv(dataset_path, sep='\t', header=None, names=["Label", "Message"])

# Step 3: Preprocessing - Check for missing data
print(f"Dataset shape: {dataset.shape}")
print(dataset.head())

# Step 4: Split the data into training and testing sets
X = dataset['Message']
y = dataset['Label']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Step 5: Vectorization - Convert text messages into numerical data (TF-IDF)
vectorizer = TfidfVectorizer(stop_words='english')
X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)

# Step 6: Train the Na√Øve Bayes classifier
nb = MultinomialNB()
nb.fit(X_train_tfidf, y_train)

# Step 7: Make predictions on the test set
y_pred = nb.predict(X_test_tfidf)

# Step 8: Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.4f}")
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```

```python
import pandas as pd
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import roc_auc_score, roc_curve
import matplotlib.pyplot as plt

# Step 1: Load the dataset (Breast Cancer dataset)
data = load_breast_cancer()
X = data.data
y = data.target

# Step 2: Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Step 3: Train the Na√Øve Bayes classifier
nb_classifier = GaussianNB()
nb_classifier.fit(X_train, y_train)

# Step 4: Make predictions on the test set (probabilities for ROC curve)
y_probs = nb_classifier.predict_proba(X_test)[:, 1]  # Probability estimates for the positive class

# Step 5: Calculate the ROC-AUC score
roc_auc = roc_auc_score(y_test, y_probs)
print(f"ROC-AUC Score: {roc_auc:.4f}")

# Step 6: Plot the ROC curve
fpr, tpr, thresholds = roc_curve(y_test, y_probs)

plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='b', label=f'ROC curve (AUC = {roc_auc:.4f})')
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc='lower right')
plt.show()
```

##Q46:  Write a Python program to train an SVM Classifier and visualize the Precision-Recall Curve.

```python
import pandas as pd
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import roc_auc_score, roc_curve, precision_recall_curve # import precision_recall_curve
import matplotlib.pyplot as plt

# Load the Breast Cancer dataset
data = load_breast_cancer()
X = data.data
y = data.target

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train the Gaussian Naive Bayes classifier
nb_classifier = GaussianNB()
nb_classifier.fit(X_train, y_train)

# Make predictions on the test set
y_pred = nb_classifier.predict(X_test) # Predict classes instead of probabilities

# Compute Precision-Recall curve
precision, recall, _ = precision_recall_curve(y_test, y_pred) # Use y_pred with class predictions

plt.plot(recall, precision, marker='.')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.show()
```
