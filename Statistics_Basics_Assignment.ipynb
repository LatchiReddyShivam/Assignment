##Q1. Explain the different types of data (qualitative and quantitative) and provide examples of each. Discuss nominal, ordinal, interval, and ratio scales.




##Ans-Data can be broadly categorized into two types: **qualitative (categorical)** and **quantitative (numerical)**.



*   Qualitative Data: Describes categories or qualities and cannot be measured numerically.
 *   Example: Colors of cars (red, blue, black), types of fruits (apple, banana).
 *   Includes:
 *   **Nominal Scale:** Categories without any order (e.g., blood type – A, B, AB, O).
 *   **Ordinal Scale:** Categories with a logical order (e.g., education level – high school, undergraduate, postgraduate).


*   Quantitative Data: Consists of numbers and represents measurable quantities.
 *   Example: Age, height, weight.
 *   Includes:
 *   **Interval Scale:** Numerical data without a true zero (e.g., temperature in Celsius).
 *   **Ratio Scale:** Numerical data with a true zero, allowing for the calculation of ratios (e.g., height in cm, weight in kg).






##Q2. What are the measures of central tendency, and when should you use each? Discuss the mean, median, and mode with examples and situations where each is appropriate.



##Ans-The measures of central tendency summarize a dataset with a single value that represents the center of the data.




*   **Mean:** The arithmetic average. Best used with **symmetric data** without outliers.
 *   Example: Average marks in a class.

*   **Median:** The middle value when data is ordered. Best used with **skewed data or data with outliers.**

 *   Example: Median income in a locality.
*   **Mode:** The value that appears most frequently. Best used for **categorical data** or identifying trends.
 *   Example: Most common shoe size sold in a store.




##Q3. Explain the concept of dispersion. How do variance and standard deviation measure the spread of data?


##Ans- Dispersion refers to the extent to which data values are spread out from the central value (mean).


*   **Variance:** Measures the average of the squared differences from the mean. It shows how much the data varies.




*   **Standard Deviation:** The square root of variance. It is in the same unit as the data, making it more interpretable.

##Higher values of variance and standard deviation indicate greater spread in the data.



##Q4. What is a box plot, and what can it tell you about the distribution of data?



##Ans-A **box plot** (or box-and-whisker plot) is a graphical representation of a dataset that displays its minimum, **first quartile (Q1)**, **median (Q2)**, **third quartile (Q3)**, and **maximum.**
##It reveals:


*   Central tendency (median)

*   Spread (interquartile range)

*   Skewness

*   Potential outliers (data points beyond 1.5 × IQR).



##Q5. Discuss the role of random sampling in making inferences about populations.



##Ans-**Random sampling** is a technique where each member of a population has an equal chance of being selected. It ensures that the sample is representative, reducing bias.
##Its role:


*   Allows for **unbiased estimates** of population parameters.

*   Enables valid **statistical inferences.**

*   Increases **generalizability** of results.



##Q6. Explain the concept of skewness and its types. How does skewness affect the interpretation of data?



##Ans-**Skewness** measures the asymmetry of a data distribution.


*   **Positive Skew (Right-skewed):** Tail on the right; mean > median.

*   **Negative Skew (Left-skewed):** Tail on the left; mean < median.

*   **Zero Skew:** Symmetrical data; mean ≈ median.

##Skewness affects how the **mean, median, and mode** are interpreted and can influence which statistical methods are appropriate.




##Q7. What is the interquartile range (IQR), and how is it used to detect outliers?




##Ans- **IQR** is the range between the first (Q1) and third quartiles (Q3):
##**IQR** = **Q3** − **Q1**
##It measures the middle 50% of the data and is resistant to outliers. Outliers are detected if they fall below:
##**Q1** − **1.5** × **IQR** or above **Q3 + 1.5 × IQR**



##Q8. Discuss the conditions under which the binomial distribution is used.




##Ans-The **binomial distribution** applies when:




##1.   There are a fixed number of trials.

##2.   Each trial has two outcomes (success/failure).

##3.   Trials are independent.

##4.   Probability of success is constant.

##Example: Tossing a coin 10 times and counting the number of heads.



##Q9. Explain the properties of the normal distribution and the empirical rule (68-95-99.7 rule).


##Ans-**Normal distribution** is a bell-shaped, symmetric curve centered at the mean.
##Properties:


*   Mean = Median = Mode.

*   Total area under the curve is 1.

*   Defined by mean (μ) and standard deviation (σ).

##**Empirical Rule:**


*   68% of data within 1σ,

*   95% within 2σ,

*   99.7% within 3σ from the mean.

##Q10. Provide a real-life example of a Poisson process and calculate the probability for a specific event.



##Ans-
##Example: Number of emails received per hour.

##Let λ = 5 emails/hour. What is the probability of receiving exactly 3 emails?

##Using Poisson formula:
##**P(X = k) = (λ^k × e^−λ) / k!**

##**P(3) = (5^3 × e^−5) / 3! = (125 × 0.00674) / 6 ≈ 0.1404**


##Q11. Explain what a random variable is and differentiate between discrete and continuous random variables.


##Ans- A random variable is a function that assigns numerical values to the outcomes of a random process.


*   Discrete Random Variable: Takes countable values (e.g., number of children).


*   Continuous Random Variable: Takes any value within a range (e.g., height, temperature).
