#Q1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.

##ANS :

```python
def sum_of_even_numbers(numbers):
    even_sum = sum([num for num in numbers if num % 2 == 0])
    return even_sum

# Example usage:
numbers = [1, 2, 3, 4, 5, 6]
print(sum_of_even_numbers(numbers))
```

#Q4. Write a Python function that checks if a given number is prime or not from 1 to 200.

##ANS :

```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

# Example usage:
for num in range(1, 201):
    if is_prime(num):
        print(f"{num} is a prime number")
```

#Q5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

##ANS :

```python
class FibonacciIterator:
    def __init__(self, max_terms):
        self.max_terms = max_terms
        self.count = 0
        self.a, self.b = 0, 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.max_terms:
            raise StopIteration
        fib = self.a
        self.a, self.b = self.b, self.a + self.b
        self.count += 1
        return fib

# Example usage:
fib_iter = FibonacciIterator(10)
for num in fib_iter:
    print(num)
```

#Q6.Write a generator function in Python that yields the powers of 2 up to a given exponent.

##ANS :

```python
def powers_of_two(exponent):
    for n in range(exponent + 1):
        yield 2 ** n

# Example usage:
for power in powers_of_two(5):
    print(power)
```

#Q7. Implement a generator function that reads a file line by line and yields each line as a string.

##ANS :

```python
def file_line_reader(file_path):
    try:
        with open(file_path, 'r') as file:
            for line in file:
                yield line.strip()
    except FileNotFoundError:
        print(f"Error: The file '{file_path}' was not found.")
    except IOError:
        print(f"Error: An issue occurred while reading the file '{file_path}'.")

# Example usage
file_path = 'example.txt'  # Provide the correct file path
for line in file_line_reader(file_path):
    print(line)
```

#Q8. Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.

##ANS :

```python
def sort_by_second_element(tuples_list):
    return sorted(tuples_list, key=lambda x: x[1])

# Example usage:
tuples_list = [(1, 4), (2, 5), (4, 6), (6, 8)]
sorted_list = sort_by_second_element(tuples_list)
print(sorted_list)
```

#Q9. Write a Python program that uses ***'map()'*** to convert a list of temperatures from Celsius to Fahrenheit.

##ANS :

```python
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

# Example usage:
celsius_temps = [0, 10, 20, 30, 40]
fahrenheit_temps = list(map(celsius_to_fahrenheit, celsius_temps))
print(fahrenheit_temps)
```

#Q10. Create a Python program that uses ***'filter()'*** to remove all the vowels from a given string.

##ANS :

```python
def remove_vowels(s):
    vowels = 'aeiouAEIOU'
    return ''.join(filter(lambda x: x not in vowels, s))

# Example usage:
string = "hello world"
no_vowel_string = remove_vowels(string)
print(no_vowel_string)
```

#Write a Python program, which returns a list with 2-tuples. Each tuple consists of the order number and the product of the price per item and the quantity. The product should be increased by 10,- € if the value of the order is smaller than 100.00 €.
#Write a Python program using lambda and map.

##ANS : To solve this accounting routine problem, we need to process the list of book orders. The task requires us to use lambda and map to generate a new list with 2-tuples where each tuple contains the order number and the product of price per item and quantity. Additionally, if the product of the price and quantity is less than " 100.00 € ", we add  " 10.00 € " to it.

```python
# The list of book orders
book_orders = [
    [34587, "Learning Python, Mark Lutz", 4, 40.95],
    [98762, "Programming Python, Mark Lutz", 5, 56.80],
    [77226, "Head First Python, Paul Barry", 3, 32.95],
    [88112, "Einführung in Python3, Bernd Klein", 3, 24.99]
]

# Function to calculate the total cost of each order
# If the total cost is below 100.00 €, add 10.00 € to it.
calculate_order_total = lambda order: (
    order[0],  # Order number
    order[2] * order[3] + (10 if order[2] * order[3] < 100 else 0)  # Quantity * Price per item with 10 € increase if < 100 €
)

# Apply the function to each order using map()
order_totals = list(map(calculate_order_total, book_orders))

# Output the result
print(order_totals)
```

#Explanation of the Code:

1. **Input Data (book_orders)**:This is a list containing sublists. Each sublist represents an order:

*  "34587": Order number

*   "Learning Python, Mark Lutz": Book title and author
*   "4": Quantity of the book ordered

*  "40.95": Price per item

2. **Lambda Function** :
*   For each order, the lambda function calculates the total price by multiplying " quantity * price_per_item ".

*  If the total price is less than "100.00 €", "10.00 €" is added to the total.
*  The lambda function returns a tuple with the order number and the final total price.

3.   Using ***'map()'*** :

*   ***'map()'*** is used to apply the lambda function (calculate_order_total) to each order in the book_orders list.
*  The result of ***'map()'*** is converted to a list of tuples.

4. **Output** : The final output is a list of tuples, where each tuple contains the order number and the final total cost.

#The Output represents :

* The first order ***(34587)*** has a total of 40.95 * 4 = **" 163.80 € "**, which is above 100 €, so no extra charge is added.

* The second order ***(98762)*** has a total of 56.80 * 5 = **" 284.00 € "**, also no extra charge.
*   The third order ***(77226)*** has a total of 32.95 * 3 = 98.85 €, which is below 100 €, so 10.00 € is added, resulting in 98.85 + 10 = **" 108.85 € "**.

*  The fourth order ***(88112)*** has a total of 24.99 * 3 = 74.97 €, so 10.00 € is added, resulting in **" 84.97 € "**.
